```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Sneha Vijay\Downloads\virtual intern\meriskill\Sales Data.csv")
```

```python
# Convert 'Order Date' to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Extract 'Month' and 'Hour' from 'Order Date'
data['Month'] = data['Order Date'].dt.month
data['Hour'] = data['Order Date'].dt.hour
```

```python
monthly_sales = data.groupby('Month')['Sales'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
city_sales = data.groupby('City')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=city_sales.values, y=city_sales.index, orient='h')
plt.xlabel('Total Sales')
plt.ylabel('City')
plt.title('Sales by City')
plt.show()
```

```python
hourly_orders = data['Hour'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=hourly_orders.index, y=hourly_orders.values)
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.title('Hourly Order Distribution')
plt.show()
```

```python
# Top-selling products
top_products = data['Product'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.values, y=top_products.index, orient='h')
plt.xlabel('Number of Units Sold')
plt.ylabel('Product')
plt.title('Top-selling Products')
plt.show()
```

```python

# Total revenue by product
product_revenue = data.groupby('Product')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=product_revenue.values, y=product_revenue.index, orient='h')
plt.xlabel('Total Sales')
plt.ylabel('Product')
plt.title('Total Sales by Product')
plt.show()
```

```python
# Hourly sales analysis
hourly_sales = data.groupby('Hour')['Sales'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=hourly_sales.index, y=hourly_sales.values)
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales')
plt.title('Hourly Sales Trends')
plt.show()

```

```python
# Monthly sales analysis
monthly_sales = data.groupby('Month')['Sales'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Analysis')
plt.show()
```

```python
# Order quantity distribution
plt.figure(figsize=(8, 6))
sns.histplot(data['Quantity Ordered'], kde=True)
plt.xlabel('Quantity Ordered')
plt.ylabel('Frequency')
plt.title('Order Quantity Distribution')
plt.show()
```

```python

# Average order quantity
average_quantity = data['Quantity Ordered'].mean()
print(f'Average Order Quantity: {average_quantity:.2f}')
```
